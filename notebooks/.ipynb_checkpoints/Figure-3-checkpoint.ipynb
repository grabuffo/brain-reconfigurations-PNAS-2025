### Extract mouse triplet combinations 

```python
from collections import Counter
import numpy as np
c_pre = data_pre['combinations']
c_post = data_post['combinations']

def find_common_elements_and_unique_indices(c_pre, c_post):
    # Convert the arrays to tuples for easier comparison
    pre_elements = [(item[0], tuple(item[1])) for item in c_pre]
    post_elements = [(item[0], tuple(item[1])) for item in c_post]
    
    # Count occurrences in both lists
    pre_counter = Counter(pre_elements)
    post_counter = Counter(post_elements)
    
    # Find common elements with their minimum counts
    common_elements_counts = pre_counter & post_counter
    common_elements = list(common_elements_counts.elements())
    
    # Find unique indices in the original lists
    used_pre_indices = set()
    used_post_indices = set()
    indices_in_pre = []
    indices_in_post = []
    
    for element in common_elements:
        # Find the next unused index in c_pre for this element
        for i, x in enumerate(pre_elements):
            if x == element and i not in used_pre_indices:
                indices_in_pre.append(i)
                used_pre_indices.add(i)
                break
        
        # Find the next unused index in c_post for this element
        for i, x in enumerate(post_elements):
            if x == element and i not in used_post_indices:
                indices_in_post.append(i)
                used_post_indices.add(i)
                break
    
    return common_elements, indices_in_pre, indices_in_post

common_elements, indices_in_pre, indices_in_post = find_common_elements_and_unique_indices(c_pre, c_post)
```

### Scalar measures for mice triplets

```python
group='RSC'

NDP_s_pre=np.zeros(len(common_elements))
NDP_s_post=np.zeros(len(common_elements))
NDP_e_pre=np.zeros(len(common_elements))
NDP_e_post=np.zeros(len(common_elements))

FC_s_pre=np.zeros(len(common_elements))
FC_s_post=np.zeros(len(common_elements))
FC_e_pre=np.zeros(len(common_elements))
FC_e_post=np.zeros(len(common_elements))

FCD_s_pre=np.zeros(len(common_elements))
FCD_s_post=np.zeros(len(common_elements))
FCD_e_pre=np.zeros(len(common_elements))
FCD_e_post=np.zeros(len(common_elements))

for triplet_id in range(len(common_elements)):
    group=common_elements[triplet_id][0]
    idx=common_elements[triplet_id][1]

    ########################################
    # Statistics for simulated mice triplets
    bold_s_pre = Bold_sim_pre[indices_in_pre[triplet_id]]
    bold_s_pre = stats.zscore(bold_s_pre[:,75:],axis=1)
    
    NDP_s_pre[triplet_id]=np.max(np.mean(bold_s_pre,axis=0)**2)
    FC_s_pre[triplet_id]=np.mean(np.corrcoef(bold_s_pre)[np.triu_indices(mregions,1)]) 
    fcd=np.corrcoef(go_edge(bold_s_pre.T))
    FCD_s_pre[triplet_id]=np.var(fcd[np.triu_indices(len(fcd),1)])

    bold_s_post = Bold_sim_post[indices_in_post[triplet_id]]
    bold_s_post = stats.zscore(bold_s_post[:,75:],axis=1)
    
    NDP_s_post[triplet_id]=np.max(np.mean(bold_s_post,axis=0)**2)
    FC_s_post[triplet_id]=np.mean(np.corrcoef(bold_s_post)[np.triu_indices(mregions,1)]) 
    fcd=np.corrcoef(go_edge(bold_s_post.T))
    FCD_s_post[triplet_id]=np.var(fcd[np.triu_indices(len(fcd),1)])

    ########################################
    # Statistics for empirical mice triplets
    for i in range(3):
        NDP_e_pre[triplet_id]+=np.max(NDP[group]['%d_pre'%idx[i]])
        FC_e_pre[triplet_id]+=np.mean(FCs[group]['%d_pre'%idx[i]][np.triu_indices(mregions,1)]) 
        FCD_e_pre[triplet_id]+=np.var(dFCs[group]['%d_pre'%idx[i]][np.triu_indices(len(dFCs[group]['%d_pre'%idx[i]]),1)])
        
        NDP_e_post[triplet_id]+=np.max(NDP[group]['%d_post'%idx[i]])
        FC_e_post[triplet_id]+=np.mean(FCs[group]['%d_post'%idx[i]][np.triu_indices(mregions,1)]) 
        FCD_e_post[triplet_id]+=np.var(dFCs[group]['%d_post'%idx[i]][np.triu_indices(len(dFCs[group]['%d_post'%idx[i]]),1)])
        
    NDP_e_pre[triplet_id]/=3    
    NDP_e_post[triplet_id]/=3  
    FC_e_pre[triplet_id]/=3    
    FC_e_post[triplet_id]/=3  
    FCD_e_pre[triplet_id]/=3    
    FCD_e_post[triplet_id]/=3     
```

### FC, dFC for specific triplet 116 (best fit)

```python
triplet_id=116
group=c_pre[triplet_id][0]
idx=c_pre[triplet_id][1]

# Simulated
bold_s = Bold_sim_pre[triplet_id]
bold_s = stats.zscore(bold_s[:,75:],axis=1)

FC_s = np.corrcoef(bold_s)
FC_s_flat=FC_s[np.triu_indices(mregions,1)]

FCD_s= np.corrcoef(go_edge(bold_s[:,75:].T))
FCD_s_flat=FCD_s[np.triu_indices(len(FCD_s),1)]

# Empirical
FC_e_triplet=np.zeros((3,mregions,mregions))
for i in range(3):
    FC_e_triplet[i,:,:]=FCs[group]['%d_pre'%idx[i]]
FC_e_flat=np.mean(FC_e_triplet,axis=0)[np.triu_indices(mregions,1)]

FCD_e_triplet=np.zeros((3,len(NDP[group]['%d_pre'%idx[i]]),len(NDP[group]['%d_pre'%idx[i]])))
for i in range(3):
    FCD_e_triplet[i,:,:]=dFCs[group]['%d_pre'%idx[i]]
FCD_e_flat = extract_and_concatenate_upper_triangular(FCD_e_triplet)
```

```python
ks_dist_fc = stats.ks_2samp(FC_s_flat,FC_e_flat)
ks_dist_fcd = stats.ks_2samp(FCD_s_flat,FCD_e_flat)

print(triplet_id,ks_dist_fc[0],ks_dist_fcd[0],np.corrcoef(FC_s_flat,FC_e_flat)[0,1])
```

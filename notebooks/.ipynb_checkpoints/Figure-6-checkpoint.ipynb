# Single Eta Inference

```python
ETA_PRE={}
ETA_POST={}
G_PRE={}
G_POST={}
for jjj, iii in enumerate(['CTRL','RSC','ACA','Th']):
    with open(join("../data/SBI/1eta_1G_Inference/G_eta_%s_/0"%iii, "eta_%s_reduced.pickle"%iii), "rb") as f:
        eta_iii = pickle.load(f)
    with open(join("../data/SBI/1eta_1G_Inference/G_eta_%s_/0"%iii, "g_%s_reduced.pickle"%iii), "rb") as f:
        g_iii = pickle.load(f)
    ETA_PRE[iii]= eta_iii['pre']
    ETA_POST[iii]= eta_iii['post']
    G_PRE[iii]= g_iii['pre']
    G_POST[iii]= g_iii['post']
```

```python
plt.figure(figsize=(2.,10))
for jjj, iii in enumerate(['CTRL','RSC','ACA','Th']):
    plt.subplot(4,1,jjj+1)
    plt.title(' \n ')
    A=ETA_PRE[iii]
    B=ETA_POST[iii]
    box_plot_data=[A,B]
    for i in range(len(A)):
        pr=A[i]
        ps=B[i]
        plt.scatter(1,pr,color=CL[iii],alpha=1.)
        plt.scatter(2,ps,color=CL[iii],alpha=0.4)
        if pr-ps>0:
            plt.plot([1,2],[pr,ps],color='darkgray',alpha=0.6)
        else:
            plt.plot([1,2],[pr,ps],color='darkgray',alpha=0.6,linestyle='--')
            
    plt.xlim((0.75,2.25))
    plt.xticks([1,2],['pre','post'],fontsize=12)
    statistic, p = scipy.stats.wilcoxon(A, B, alternative='less')
    print(statistic)
    bottom, top = np.min(np.concatenate([A,B])),np.max(np.concatenate([A,B]))
    bars(p, bottom, top,0.1)
    plt.ylim((bottom-np.abs(bottom)/40, top+np.abs(top)/9))
    plt.ylabel(r'$\eta_{{{}}}$'.format(et[jjj]), fontsize=12)
plt.tight_layout()
plt.savefig('../figures/Figure6/Inference_eta.png',dpi=300)
plt.show()
```

```python
plt.figure(figsize=(2.,10))
for jjj, iii in enumerate(['CTRL','RSC','ACA','Th']):
    plt.subplot(4,1,jjj+1)
    plt.title(' \n ')
    A=G_PRE[iii]
    B=G_POST[iii]
    box_plot_data=[A,B]
    for i in range(len(A)):
        pr=A[i]
        ps=B[i]
        plt.scatter(1,pr,color=CL[iii],alpha=1.)
        plt.scatter(2,ps,color=CL[iii],alpha=0.4)
        if pr-ps>0:
            plt.plot([1,2],[pr,ps],color='darkgray',alpha=0.6)
        else:
            plt.plot([1,2],[pr,ps],color='darkgray',alpha=0.6,linestyle='--')
            
    plt.xlim((0.75,2.25))
    plt.xticks([1,2],['pre','post'],fontsize=12)
    statistic, p = scipy.stats.wilcoxon(A, B, alternative='greater')
    bottom, top = np.min(np.concatenate([A,B])),np.max(np.concatenate([A,B]))
    bars(p, bottom, top,0.002)
    plt.ylim((bottom-np.abs(bottom)/80, top+np.abs(top)/50))
    #if jjj==0:
    plt.ylabel('$G$', fontsize=12)
    #plt.ylim((bottom-0.01, top+0.01))
plt.tight_layout()
plt.savefig('../figures/Figure6/Inference_g.png',dpi=300)
plt.show()
```

```python
# read csv file and drop 
df = pd.read_csv(data_path+"df235_prepost2.csv", sep='\t', encoding='utf-8')
df.head(6)
```

```python
fig, ax = plt.subplots(4, 6, figsize=(6, 4.5), sharey="row")
groups = ["Th", "ACA", "RSC", "CTRL"]
columns = ['eta_DMN', "eta_Vis", "eta_LCN", "eta_BF", "eta_HPF", "eta_Th"]

for i in range(4):
    group = groups[i]
    subs = [2, 3, 5]
    palette=[CL[group],".85"]
    for col in range(6):
        # filter dataframe for group and each columns[col]
        
        df_ = df[df["group"] == group][df["n_subj"] == 3]
        
        #statistics
        x=np.asarray(df[df["group"] == group][df["n_subj"] == 3][df['prepost']=='pre'][columns[col]])
        y=np.asarray(df[df["group"] == group][df["n_subj"] == 3][df['prepost']=='post'][columns[col]])
        p=stats.wilcoxon(x,y)[1]
        
        sns.violinplot(x="n_subj", y=columns[col], 
                       data=df_, palette=palette, gridsize=50,
                       hue="prepost", split=True, alpha=0.85, ax=ax[i, col])             
            
        if p<0.0001:
            ax[i, col].legend(labels=[], title='*', loc="lower right",frameon=False,title_fontsize=12)
        elif p<0.001:
            ax[i, col].legend(labels=[], title='', loc="lower right",frameon=False,title_fontsize=12)
        elif p<0.01:
            ax[i, col].legend(labels=[], title='', loc="lower right",frameon=False,title_fontsize=12)
        elif p<0.05:
            ax[i, col].legend(labels=[], title='', loc="lower right",frameon=False,title_fontsize=12)
        else:
            ax[i, col].legend(labels=[], title='', loc="center right",frameon=False,title_fontsize=12)
            
            
        #ax[i, col].set_ylim(-6,-3.5)
        if col == 0:
            ax[i, col].set_ylabel(f"{group}", fontsize=12)
        else: 
            ax[i, col].set_ylabel("")
        if i == 0:
            ax[i, col].set_title(r"$\eta$ "+ columns[col].strip('eta_'), fontsize=12) # fontweight="bold"
        
        if i == 3:
            ax[i, col].set_xlabel("")#"Number of subjects")
        else:
            ax[i, col].set_xlabel("")
        ax[i, col].set_xticks([])
plt.tight_layout()
plt.savefig(join("../figures/Figure6/PNAS_posterior_peaks.pdf"), bbox_inches="tight")
```

```python
fig, ax = plt.subplots(4, 1, figsize=(1.35, 4.5), sharey="row")
groups = ["Th", "ACA", "RSC", "CTRL"]
columns = ['eta_DMN', "eta_Vis", "eta_LCN", "eta_BF", "eta_HPF", "eta_Th","g"]
#palette = ['royalblue', 'red']

for i in range(4):
    group = groups[i]
    subs = [2, 3, 5]
    palette=[CL[group],".85"]
        
    df_ = df[df["group"] == group][df["n_subj"] == 3]

    #statistics
    x=np.asarray(df[df["group"] == group][df["n_subj"] == 3][df['prepost']=='pre']["g"])
    y=np.asarray(df[df["group"] == group][df["n_subj"] == 3][df['prepost']=='post']["g"])
    p=stats.wilcoxon(x,y)[1]

    sns.violinplot(x="n_subj", y="g", 
                   data=df_, palette=palette, gridsize=50,  #"seismic"
                   hue="prepost", split=True, alpha=0.85, ax=ax[i])

    if p<0.0001:
        ax[i].legend(labels=[], title='*', loc="upper right",frameon=False,title_fontsize=12)
    elif p<0.001:
        ax[i].legend(labels=[], title='', loc="upper right",frameon=False,title_fontsize=12)
    elif p<0.01:
        ax[i].legend(labels=[], title='', loc="upper right",frameon=False,title_fontsize=12)
    elif p<0.05:
        ax[i].legend(labels=[], title='', loc="upper right",frameon=False,title_fontsize=12)
    else:
        ax[i].legend(labels=[], title='', loc="upper right",frameon=False,title_fontsize=12)


    ax[i].set_ylabel("")
    if i==0:
        ax[i].set_title("G", fontsize=12) # fontweight="bold"
    ax[i].set_xlabel("")#"Number of subjects")
    ax[i].set_xticks([])
plt.tight_layout()
plt.savefig(join("../figures/Figure6/PNAS_posterior_peaksG6.pdf"), bbox_inches="tight")
```
